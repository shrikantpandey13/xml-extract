```python
def data_extract(filename):
    filename = filename
    tree = ET.parse(filename)
    chemical = tree.findall('Chemical')
    data = []
    for chem in chemical:
        for ele in chem:
            if(ele.tag=="NameList"):
                name_list = ' : '.join([ele1.text for ele1 in ele.findall('SystematicName')])
#             print(name_list)
            elif(ele.tag=='NumberList'):
                cas_number = ele.find('CASRegistryNumber').text
#             print(cas_number)
            elif(ele.tag=='FormulaList'):
                molecular = ele.find('MolecularFormula').text
#             print(molecular)
        row = [name_list, cas_number, molecular]
#         print('*****************')
        data.append(row)
#         print(data)
    df = pd.DataFrame(data,columns=['CompundName','CASRegistryNumber','MolecularFormula'])
    del name_list, cas_number, molecular
    return df
    
```

```python
from datetime import datetime
start_time = datetime.now()
df = data_extract('demo.xml')
end_time = datetime.now()-start_time
print("Time elapsed(hh:mm:ss){}".format(end_time))
df.to_csv("df_large_cas.csv",index=False)
```
